Step 08 - Burgers' Equation in 2D

Here, the goal was to understand the equation that combines both 2D non-linear convection and 2D diffusion. Burgers' equation is one of the most famous equations in CFD because it has many of the same mathematical properties as the Navier–Stokes equations, but it is much simpler.

It allows us to learn: nonlinear transport, diffusion, shock formation, numerical stability, boundary conditions, without the complexity of pressure and multiple velocity components present in the NS equation.
