Step 03 - ID Diffusion Equation

Here, the goal was to understand the concept of discretization of second order deriatives using the central difference scheme (a combination of forward and backward difference of the first derivative), also the use of the Taylor expansion series. The second order derivatives arises from the 1D diffusion equation.

```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt


# initializing the parameters
nx = 41 # number of grid points in x-direction
dx = 2 / (nx - 1) # grid spacing in x-direction
nt = 20 # number of time steps
nu = 0.3 # kinematic viscosity
sigma = .2 # CFL condition
dt = sigma * dx**2 / nu # time step size


# creating the grid and setting boundary conditions
u = np.ones(nx) # initial condition: u = 1 everywhere
u[int(.5 / dx):int(1 / dx + 1)] = 2 # setting u = 2 between 0.5 and 1


#implementing the diffusion equation using finite difference method

un = np.ones(nx) # temporary array to hold the values of u at the previous time step

for n in range(nt): # loop over time steps
    un = u.copy() # copy the values of u to un
    for i in range(1, nx - 1): # loop over spatial points
        u[i] = un[i] + nu * dt / dx**2 * (un[i + 1] - 2 * un[i] + un[i - 1]) # finite difference scheme


# plotting the results

plt.plot(np.linspace(0, 2, nx), u)
```
